```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
from pandas.tools.plotting import scatter_matrix
%matplotlib inline
```

```python
# upload date set into data frame
df = pd.read_csv("/Users/OzlemCelik/Desktop/googleplaystore_3.csv", low_memory=False)
```

```python
#drop duplicate rows
df = df.drop_duplicates()
```

```python
df.dropna(how='any', inplace=True)
```

```python
df.isnull().sum()
```

<b>1)Choose one variable and plot that variable four different ways.

```python
sns.violinplot(x='Rating', data=df)
plt.xlabel('Ratings', fontsize=10)
plt.title("Rating Frequency", fontsize=10)
plt.tight_layout()
plt.show()
print('Violin plot shows the distribution of continous data across several levels.')


df['Rating'].plot(kind='box')
plt.ylabel('Rating', fontsize=10)
plt.title("Rating Frequency", fontsize=10);
plt.tight_layout()
plt.show()


sns.distplot(df['Rating'], kde=False, bins=20)
plt.ylabel('Count', fontsize=10)
plt.xlabel('Ratings', fontsize=10)
plt.title("Rating Frequency", fontsize=10)
plt.tight_layout()
plt.show()
print('Histogram plot shows distribution of rating. Able to quickly see the density of values and how much variation there is.')



sns.distplot(df['Rating'], rug=True, hist=False)
plt.xlabel('Ratings', fontsize=10)
plt.title("Rating Frequency", fontsize=10)
plt.tight_layout()
plt.show()
print('Not clear just for comparision')
```

<b>1)Choose two continuous variables, and plot them three different ways.

```python
sns.regplot(x='Reviews', y='Rating', data=df)
plt.ylabel('Rating', fontsize=10)
plt.xlabel('Reviews', fontsize=10)
plt.title("Rating vs Review", fontsize=10)
plt.tight_layout()
plt.show()


sns.jointplot("Reviews", "Rating", data=df, kind="reg", xlim=(0, 60), ylim=(0, 6), color="m", height=7)
plt.ylabel('Ratings', fontsize=10)
plt.xlabel('Reviews', fontsize=10)
plt.title("Rating vs Reviews", fontsize=10)
plt.tight_layout()
plt.show()


sns.lineplot(x="Rating", y="Reviews",data=df)
plt.ylabel('Rating', fontsize=10)
plt.xlabel('Reviews', fontsize=10)
plt.title("Rating vs Review", fontsize=10)
plt.tight_layout()
plt.show()
```

<b>3)Choose one continuous variable and one categorical variable, and plot them six different ways

```python
sns.pointplot(x='Rating', y='Type', data=df, size=4, aspect=2, join=False)
plt.show()



sns.boxplot(x='Rating', y='Type', data=df)
plt.show()



sns.boxplot(x="Rating", y="Type", data=df,whis="range", palette="vlag")
plt.show()


sns.violinplot(x='Rating', y='Type', data=df)
plt.show()


sns.barplot(x='Type', y='Rating', data=df)
plt.show()


sns.stripplot(x='Type', y='Rating', data=df, jitter=0.05)
plt.show()
```
